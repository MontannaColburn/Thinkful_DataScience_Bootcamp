# Present the same data multiple ways

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
plt.rcParams['figure.figsize'] = (20.0, 10.0)
#plt.rcParams.update({'font.size': 22})
```

```python
df=pd.read_csv("air_pollutants.csv")
df.head(5)
```

```python
#update to have the same unit of measurement 
Data= df[(df['UNIT_MEASURE'] == 'Kilograms per capita')]
Data.head(5)
```

# One variable plotted four ways 

```python
plt.subplot(1, 3, 1)
plt.ylim(0,1300)
sns.distplot(Data.OBS_VALUE, kde=False)
plt.xlabel('OBS Value')
plt.ylabel('Frequency')
plt.title('OBS Value Frequency Distribution')

#add a rug plot
plt.subplot(1,3,2)
plt.ylim(0,1300)
sns.distplot(Data.OBS_VALUE, kde=False, rug = True)
plt.xlabel('OBS Value')
plt.ylabel('Frequency')
plt.title('OBS Value Frequency Distribution with rug')

#add in bins parameter
plt.subplot(1,3,3)
plt.ylim(0,1300)
sns.distplot(Data.OBS_VALUE, bins=10, kde=False, rug = True)
plt.xlabel('OBS Value')
plt.ylabel('Frequency')
plt.title('OBS Value Frequency Distribution Bins = 10');
```

This graphic shows how the data is distributed. As we can see, the distribution is not normal with our peak to the left. 

```python
sns.boxplot(x='OBS_VALUE', data=Data)  
plt.title('OBS Value Boxplot')
```

 We can see the distribution of the data by quartiles and outliers. As we can see there is quite a lot of outliers in the data for OBS Value.

```python
sns.barplot(x="OBS_VALUE", data=Data)
plt.title('OBS Value Bar')
```

This graphic isn't really fantastic it doesn't really give us any idea on the y values.

```python
sns.stripplot(x='OBS_VALUE', data=Data, orient="h")
```

We can see how the data is spread here.

# Two continous variables plotted 3 ways 

```python
g = sns.lmplot(y='OBS_VALUE', # Variable 1.
               x='YEAR', # Variable 2.
               data=Data, # Data
               fit_reg=False, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("OBS Value")
g.set_xlabels("Year")
plt.title('Scatterplot: OBS Value by Year')
plt.show()

g = sns.lmplot(y='OBS_VALUE', 
               x='YEAR',
               data=Data, # Data.
               fit_reg=True, # The regression line also includes a 95% confidence envelope.
               scatter_kws={'alpha':0.4})
g.set_ylabels("OBS Value")
g.set_xlabels("Year")
plt.title('Scatterplot with regression line: OBS Value by Year')
plt.show()
```

```python
sns.set(style="whitegrid")

ax = sns.boxplot(x='YEAR',y='OBS_VALUE',data=Data,palette='pastel')  

plt.title('Plot 4: OBS Value by Year')

sns.despine(offset=10, trim=True)
ax.set(xlabel='Year', ylabel='OBS Value')

plt.show()
```

```python
sns.set(style="white")

g = sns.lmplot(y='OBS_VALUE', 
               x='YEAR',
               data=Data, # Data.
               col='POLLUTANT',
               hue='POLLUTANT',
               fit_reg=True,
               ci=False,
               scatter_kws={'alpha':0.4})
g.set_ylabels("OBS Value")
g.set_xlabels("Year")
plt.suptitle('Scatterplot Regression: OBS Value by Year, split by gender and time')
plt.subplots_adjust(top=0.9)
plt.show()
```

# Just some excess graphics for fun 

```python
carbonobs2013=Data.loc[(Data['YEAR'].astype(str)=='2013')&(Data['POLLUTANT']=='Carbon Monoxide'),'OBS_VALUE']
carbonobs2014=Data.loc[(Data['YEAR'].astype(str)=='2014')&(Data['POLLUTANT']=='Carbon Monoxide'),'OBS_VALUE']
sulfuroxides2013=Data.loc[(Data['YEAR'].astype(str)=='2013')&(Data['POLLUTANT']=='Sulphur Oxides'),'OBS_VALUE']
sulfuroxides2014=Data.loc[(Data['YEAR'].astype(str)=='2014')&(Data['POLLUTANT']=='Sulphur Oxides'),'OBS_VALUE']
```

```python
plt.hist(carbonobs2013, color='red',  alpha=.5,label='Carbon Monoxide 2013')
plt.hist(carbonobs2014, color='blue', alpha=.5,label='Carbon Monoxide 2014')
plt.hist(sulfuroxides2013, color='green', alpha=.5,label='Sulfur Oxides 2013')  
plt.hist(sulfuroxides2014, color='yellow', alpha=.5,label='Sulfur Oxides 2014')
plt.xlabel('OBS Value')
plt.legend(loc='upper right')
plt.title('Plot 1: OBS Value for the years 2013 and 2014 for Carbon Monoxide and Sulfur Oxides')
plt.show()
```

```python
sns.set(style="ticks")
# Tell seaborn about the structure of our data.
g = sns.FacetGrid(Data, row="YEAR", col="POLLUTANT")
# Name the plot type and the variable to be plotted using the structure.
g.map(plt.hist, "OBS_VALUE", color="steelblue",  lw=0)
# Moving the plots apart to make room for our titles.
plt.subplots_adjust(top=0.99)
# Making a more informative axis name.
g.set_axis_labels('OBS Value')
plt.suptitle('Plot 2: OBS Value for the years 2013 and 2014 for Carbon Monoxide and Sulfur Oxides')
# Removing excess lines around the plot.
sns.despine(trim=True)
plt.show()
```

```python
ax = sns.boxplot(x='POLLUTANT',y='OBS_VALUE',hue='YEAR',data=Data)  
plt.title('Plot 3: OBS Value by Pollutant and Year')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='OBS_VALUE')
ax.legend(bbox_to_anchor=(1.1, 0.98))

plt.show()
```
